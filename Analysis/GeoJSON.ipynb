```python
import requests
import json
```

```python
coordinates = requests.get('http://localhost:3000/v1/user')
coordinates = coordinates.json()
coordinates = coordinates['data']
coordinates
```

```python
# input_file=json.load(open("./data/data.json", "r", encoding="utf-8"))

geojs={
     "type": "FeatureCollection",
     "features":[
           {
                "type":"Feature",
                "geometry": {
                "type":"LineString",
                "coordinates":d["geometry"],
            },
                "properties":d,
        
         } for d in coordinates 
    ]  
 }

output_file=open("geodata.json", "w", encoding="utf-8")
json.dump(geojs, output_file)


output_file.close()
```
